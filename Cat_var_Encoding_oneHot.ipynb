# In this we see how we can make dummy variables of classes of variable with two different method 1st is using pandas (get_dummies()) and 2nd OneHotEncoding from sklearn

```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

# **So in Dataset we have to convert categorical data into numeric format for use in machine learning model for that we OneHot Encoding in which we make dummy column of every class of varible**

```python
df
```

```python
dummy_df=pd.get_dummies(df)
dummy_df
```

# Here we dummy column of sex like male and female and their value in 0 and 1

```python
#here it creates 2 dummy column but the value are same (like if male is present
# then use 1 and for female use 0)so we need to drop 1 column
df2=pd.get_dummies(df,drop_first=True)
df2
```

# 2nd Method OneHotEncoding
# convering categorical data into OneHot Endcoding of traing ML model using sklearn

```python
from sklearn.preprocessing import OneHotEncoder
# Use sparse_output instead of sparse for newer scikit-learn versions
oh_enc = OneHotEncoder(sparse_output=False)

oh_enc_arr=oh_enc.fit_transform(df[['sex','smoker','day','time']]) #double square of input in sklearn
oh_enc_arr
```

```python
dummy_df.keys()
```

```python
oh_enc_df=pd.DataFrame(oh_enc_arr,columns=['sex_Male', 'sex_Female', 'smoker_Yes',
       'smoker_No', 'day_Thur', 'day_Fri', 'day_Sat', 'day_Sun', 'time_Lunch',
       'time_Dinner'])
oh_enc_df
```
